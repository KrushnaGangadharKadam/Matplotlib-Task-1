#### 1. Install matplotlib and import pyplot as plt

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### 2. Plot a basic line chart using plt.plot()

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
y1 = np.sin(x)
y2 = np.cos(x)

plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Study Time vs Test Score')
plt.plot(x, y, 'r--')  # 'r' = red, 'o' = circle, '--' = dashed line
plt.show()
```

#### 3. Add title and labels to axes

```python
plt.plot(x, y)
plt.title("Title and Labels to axes")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

#### 4. Change line color, style and markers

```python
plt.plot(x, y, color='red', linestyle='--', marker='o', label='Sine')
plt.title("Styled Line Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

#### 5. Plot multiple lines in same chart

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
Day = [120, 135, 150, 160, 170, 200, 190]
Month = [30, 40, 35, 45, 50, 55, 60]
Year = [20, 25, 30, 35, 40, 45, 50]

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Multiple lines')

plt.plot(days, Day, label="Day", color='green', marker='o', linestyle='-')
plt.plot(days, Month, label="Month", color='blue', marker='s', linestyle='--')
plt.plot(days, Year, label="Year", color='purple', marker='^', linestyle=':')
plt.show()

```

#### 6. Add a legend to the chart

```python
plt.plot(days, Day, label="Day")
plt.plot(days, Month, label="Month")
plt.plot(days, Year, label="Year")
plt.legend(loc='upper right',shadow=True,fontsize='large')
plt.show()
```

#### 7. Save th plot as an image file

```python
plt.plot(days, Day, label="Day")
plt.plot(days, Month, label="Month")
plt.plot(days, Year, label="Year")
plt.legend(loc='upper right',shadow=True,fontsize='large')
plt.savefig("Legend 1.png")
plt.show()
```

#### 8. Plot a bar chart using plt.bar() 

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 4]
plt.bar(categories, values, color='orange')
plt.title("Bar Chart")
plt.show()
```

#### 9. Plot a histogram using plt.hist() 

```python
exam_scores = [72, 88, 91, 85, 78, 95, 67, 82, 89, 76, 92, 85, 80, 73, 87]
plt.hist(exam_scores, bins=7, color='skyblue', edgecolor='black', rwidth=0.9)

plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Distribution of Exam Scores')
plt.show()

```

#### 10. Plot a pie chart using plt.pie() 

```python
sizes = [20, 30, 25, 25]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart")
plt.axis('equal')  
plt.show()
```

#### 15. Add annotations to plot points

```python
plt.scatter([5], [np.sin(5)], color='red')
plt.annotate("Peak", xy=(5, np.sin(5)), xytext=(6, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.plot(x, y1)
plt.title("Annotated Plot")
plt.show()
```

#### 16. Create a stacked bar chart 

```python
men = [3, 4, 5, 2]
women = [4, 3, 2, 4]
plt.bar(categories, men, label='Men')
plt.bar(categories, women, bottom=men, label='Women')
plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```

#### 17. Change background and axis colors 

```python
fig, ax = plt.subplots()
fig.patch.set_facecolor('lightgray')
ax.set_facecolor('whitesmoke')
ax.plot(x, y1, color='blue')
ax.tick_params(colors='red')
ax.spines['bottom'].set_color('green')
plt.title("Custom Colors", color='purple')
plt.show()
```

#### 18. Create a plot with twin axes (twinx/twiny)

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-', label='Sine')
ax2.plot(x, y2, 'b--', label='Cosine')
ax1.set_xlabel("X axis")
ax1.set_ylabel("Sine", color='g')
ax2.set_ylabel("Cosine", color='b')
plt.title("Twin Axes")
plt.show()
```
